```python
import numpy as np


from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier


# 参数
n_classes = 3

# 加载测试数据
iris = load_iris()
#print (iris.target)
#print (iris.data[:, pair])
#print (iris.data[:])
for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3], [1, 2], [1, 3], [2, 3]]):
   #print ("-------- begin -------------")
   #print (iris.data[:])
   #print (type(iris.data[:]))
   #print ("------- pair begin ------------")
   # 获取每行前几列
   x = iris.data[:, pair]
   y = iris.target
   # 训练
   clf = DecisionTreeClassifier().fit(x, y)
   y_pre = clf.predict(x)
   print ("----------- begin --------------")
   print (y_pre)
   print (iris.target)
   # 对应的正确率
   print (np.mean(y_pre == iris.target))
   print ("----------- end -------------")
   #iris.data[:, pair]
   #print (iris.data[:,pair])
   #print ("-------- end ------------")
```
